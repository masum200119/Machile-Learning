```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df1 = pd.read_csv(r'C:\Users\MASUM\Polynomial Regression\linear_data.csv')
df1.head()
```

```python
x_linear=df1[['x']]
y_linear=df1[['y']]
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1,2,1)
plt.scatter(x_linear, y_linear, color='blue', label='Data with Linear Relationship')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Relationship')
plt.legend()
plt.subplot(1, 2, 2)
plt.scatter(x_nonlinear, y_nonlinear, color='green', label='Data with Non-linear Relationship')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Non-linear Relationship')
plt.legend()

plt.tight_layout()
plt.show()
```
