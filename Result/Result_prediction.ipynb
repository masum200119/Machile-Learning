```python
import pandas as pd
```

```python
import numpy as np
```

```python
df=pd.read_csv('Dataset.csv')
```

```python
df
```

```python
df = df.drop(columns=['Timestamp'])
```

```python
df
```

```python
df = df.drop(columns=['Class+CT+Assignment'])
```

```python
df
```

```python
df['Relation'] = df['Relation'].map({'Yes': 1, 'No': 0})
```

```python
df
```

```python
# Mapping 'Yes' to 1, 'No' to 0, and 'sometimes' to 0.5 in the 'Study without exam' column
df['Study without exam'] = df['Study without exam'].map({'Yes': 1, 'No': 0, 'Sometimes': 2})
```

```python
df
```

```python
X = df.drop(columns=['Result(2.2)'])  # Dropping the target column
y = df['Result(2.2)']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X, y)
```

```python
input_data = {
        'Result (1.1)': [3.05],  # Example value
    'Result(1.2)': [3.26],   # Example value
    'Result(2.1)': [3.45],   # Example value
    'Tuition': [1],      # Example value
    'Relation': [1],        # Assuming '1' for 'Yes'
    'Extracurricular': [0], # Example value
    'Social media': [4],    # Example value
    'Exam_study_time': [8], # Example value
    'Study without exam': [1], # Assuming '2' for 'sometimes'
    'AVG': [3.27]         # Example value         # Example value
}

input_df = pd.DataFrame(input_data)

```

```python
prediction = model.predict(input_df)

# Output the prediction
print("Predicted Result(2.2):", prediction[0])
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr.fit(X,y)
```

```python
# Making a prediction
prediction = lr.predict(input_df)

# Output the prediction
print("Predicted Result(2.2):", prediction[0])
```

```python
from sklearn.neighbors import KNeighborsRegressor

knn = KNeighborsRegressor(n_neighbors=5)  

knn.fit(X, y)

prediction = knn.predict(input_df)

# Output the prediction
print("Predicted Result(2.2):", prediction[0])
```

```python
from sklearn.tree import DecisionTreeRegressor

# Initialize and train the model
tree = DecisionTreeRegressor(random_state=0)
tree.fit(X, y)

# Make a prediction
prediction_tree = tree.predict(input_df)
print("Decision Tree Regression Prediction:", prediction_tree[0])
```

```python
from sklearn.svm import SVR

# Initialize and train the model
svr = SVR(kernel='rbf')
svr.fit(X, y.ravel())

# Make a prediction
prediction_svr = svr.predict(input_df)
print("Support Vector Regression Prediction:", prediction_svr[0])
```

```python
from sklearn.ensemble import GradientBoostingRegressor

# Initialize and train the model
gbr = GradientBoostingRegressor(n_estimators=100, random_state=0)
gbr.fit(X, y)

# Make a prediction
prediction_gbr = gbr.predict(input_df)
print("Gradient Boosting Regression Prediction:", prediction_gbr[0])
```

```python
from sklearn.ensemble import AdaBoostRegressor

# Initialize and train the model
abr = AdaBoostRegressor(n_estimators=100, random_state=0)
abr.fit(X, y)

# Make a prediction
prediction_abr = abr.predict(input_df)
print("AdaBoost Regression Prediction:", prediction_abr[0])
```

```python
from xgboost import XGBRegressor

# Initialize and train the model
xgb = XGBRegressor(n_estimators=100, random_state=0)
xgb.fit(X, y)

# Make a prediction
prediction_xgb = xgb.predict(input_df)
print("XGBoost Regression Prediction:", prediction_xgb[0])
```

```python
from catboost import CatBoostRegressor

# Initialize and train the model
catboost = CatBoostRegressor(iterations=100, random_state=0, verbose=0)
catboost.fit(X, y)

# Make a prediction
prediction_catboost = catboost.predict(input_df)
print("CatBoost Regression Prediction:", prediction_catboost[0])
```
